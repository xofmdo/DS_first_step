### Задача 1

Постройте линейный график динамики объема продаж всех игр жанра "Sports" в Японии. Задайте графику размер 12 на 6

```python
x = df.query('Genre=="Sports"').groupby('Year')['JP_Sales'].sum().index
y = df.query('Genre=="Sports"').groupby('Year')['JP_Sales'].sum()


fig, ax = plt.subplots(figsize=(12, 6))


plt.plot(x,y, 
         color='green'
         )
plt.grid(True)

# Настройка осей
plt.xlabel('Год', fontsize=18, )
plt.ylabel('Объем продаж игр, млн. копий',  fontsize=18)
plt.title('Динамики объема продаж всех игр жанра "Sports" в Японии', fontsize=20, pad=20)
plt.margins(0.01)
plt.xticks(size=16)
plt.yticks(size=14)


plt.fill_between(x, y, color='lightblue', alpha=0.5) # заполняю пространство под линией


plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
fig, ax = plt.subplots(figsize=(12, 10))
ax.grid(color='grey', linestyle='--', linewidth=1) # задаю разметку
color = ['red', 'green', 'navy', 'orange']

plt.margins(0.01)
plt.xticks(size=16)
plt.yticks(size=14)
plt.xlabel('Год', fontsize=18, )
plt.ylabel('Объем продаж игр, млн. копий',  fontsize=18)
plt.title('Динамики продаж продаж студии Activision', fontsize=20, pad=20)

for i, value in enumerate(['NA_Sales', 'EU_Sales', 'JP_Sales','Global_Sales']):
    x = df.query('Publisher=="Activision"').groupby('Year')[value].sum().index
    y = df.query('Publisher=="Activision"').groupby('Year')[value].sum()

    plt.plot(x,y, color=color[i], label=value)
    
    if value == 'JP_Sales':
        plt.fill_between(x, y, color='lightblue', alpha=1) # даже это не выделило толком долю Японии

ax.legend(loc=2, fontsize=14)

plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
color = ['red', 'green', 'navy', 'orange']
data = [(0,0), (0,1), (1,0), (1,1)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

fig.suptitle('Динамики продаж студии "Activision"', y=0.97, color='black', size=16)


for i, value in enumerate(['NA_Sales', 'EU_Sales', 'JP_Sales','Global_Sales']):
    x = df.query('Publisher=="Activision"').groupby('Year')[value].sum().index
    y = df.query('Publisher=="Activision"').groupby('Year')[value].sum()
    ax = axes[data[i][0]][data[i][1]]
    ax.plot(x,y, color=color[i], label=value)
    ax.fill_between(x, y, color=color[::-1][i], alpha=0.3)
    ax.set_title(f'Продажи в регионе - {value}', pad=10, color='black')
    if i in [0,2]:
        ax.set_ylabel('Объем продаж игр, млн. копий', color='black')
    if i in [2,3]:
        ax.set_xlabel('Год', color='black')
    
    ax.margins(0.01)
    ax.grid(color='grey', linestyle='--', linewidth=1)


plt.show()

```

### Задача 6

Определите 5 платформ, для которых в Японии было продано больше всего игр за все время. Проиллюстрируйте величину продаж на столбчатой диаграмме, столбец с самым высоким значением окрасьте зеленым, а с самым низким - красным

```python
# подгружаю цвета и удаляю красный и зеленый, если есть
cmap = plt.get_cmap('tab20')
color = [cmap(i) for i in range(20)]
try:
    color.remove((1.0, 0.0, 0.0, 1.0)) # красный цвет RGBA
    color.remove((0.0, 0.5, 0.0, 1.0)) # зеленый цвет RGBA
except:
    pass

df1 = df.groupby('Publisher').sum()['JP_Sales'].sort_values(ascending=False).head(5)

fig, ax = plt.subplots(figsize=(12, 7))
n = ax.bar(df1.index, df1, linewidth=1.5)

# Нахожу максимальное и минимальное значение объема продаж
max_value = df1.max()
min_value = df1.min()

# Цвет самого высокого столбца - зеленый, самого низкого - красный, остальных - серый
colors = ['green' if v == max_value else 'red' if v == min_value else color.pop() for v in df1]

# Устанавливаю цвета столбцов
for i in range(len(n)):
    n[i].set_color(colors[i])
    n[i].set_edgecolor('black') # внутри bar почему-то не работает установка цвета контура

plt.grid(True)

# Настройка осей
plt.xlabel('Издатели', fontsize=18)
plt.ylabel('Объем продаж игр, млн. копий',  fontsize=18)
plt.title('Распределения объема продаж по издателям в Японии ', fontsize=20, pad=20)
plt.xticks(size=14, rotation=45)
plt.yticks(size=14)
ax.set_ylim(50, 500)
ax.bar_label(ax.containers[0], padding=3, size=14, rotation=0)

plt.show()
```
