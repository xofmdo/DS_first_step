```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузка данных

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объёма продаж всех игр жанра «Sports» в Японии. Задайте графику размер 12 на 6.

```python
df1 = df.query("Genre == 'Sports'").groupby('Year').count()

fig, ax = plt.subplots(figsize=(12,6))
sns.lineplot(data=df1,
             x="Year",
             y="JP_Sales",
             ax=ax
             )
plt.grid(True)

ax.set_xlabel('Год', size=12)
ax.set_ylabel('Объем продаж, млн.', size=12)
ax.set_title('Динамики объёма продаж всех игр жанра «Sports» в Японии', size=16, pad=18)
plt.tight_layout()  # Улучшение визуального оформления графиков
plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

```python
df_activision = df[df.Publisher == 'Activision']
df_activision = df_activision.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales']].sum()


fig, ax = plt.subplots(figsize=(12,6))

# Построение линейных графиков для каждого региона
sns.lineplot(data=df_activision, x='Year', y='NA_Sales', label='Северная Америка', color='b', linewidth=2)
sns.lineplot(data=df_activision, x='Year', y='EU_Sales', label='Европа', color='g', linewidth=2)
sns.lineplot(data=df_activision, x='Year', y='JP_Sales', label='Япония', color='r',linewidth=2)
sns.lineplot(data=df_activision, x='Year', y='Global_Sales', label='Весь мир', color='m', linewidth=2)

sns.set_style('dark')
plt.grid(True)

ax.set_title('Динамика продаж студии Activision', size=16, pad=18)
ax.set_xlabel('Год', size=12)
ax.set_ylabel('Объем продаж, млн.', size=12)
plt.legend(loc='best', frameon=False) # отключает рамку для легенды

plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

```python
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales']
df_activision = df[df['Publisher'] == 'Activision'].groupby('Year')[regions].sum()
titles = [' Северной Америке', ' Европе', ' Японии', 'о всем мире']


fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(12,18))
colors = ['r', 'g', 'b', 'y']

# Создание цикла для построения графиков
for i, ax in enumerate(axes.flatten()):
    sns.lineplot(data=df_activision, x='Year', y=regions[i], linewidth=2, ax=ax, color=colors[i])
    ax.set_title(f'Динамика продаж студии Activision в{titles[i]}', fontsize=14, pad=17)
    ax.set_xlabel('Год', fontsize=12)
    ax.set_ylabel('Объем продаж, млн.', fontsize=12)
    ax.grid(True)


plt.tight_layout()  # Улучшение визуального оформления графиков
plt.show()
```

### Задача 5

В одной координатной сетке постройте 2 гистограммы распредения мировых продаж игр издателей «Microsoft Game Studios» и «Take-Two Interactive» после 2010 года (включая 2010 год). Графики сделайте полупрозрачными, отсечки должны быть общими и соответствовать ширине столбцов обоих графиков.

```python
df_filtered = df[(df['Publisher'] == 'Microsoft Game Studios') | (df['Publisher'] == 'Take-Two Interactive')]
df_filtered = df_filtered[df_filtered['Year'] >= 2010]

fig, ax = plt.subplots(figsize=(10,6))

sns.histplot(data=df_filtered, x='Global_Sales', hue='Publisher', bins=30, alpha=0.5, ax=ax)

ax.set_title('Распредения мировых продаж игр издателей\n «Microsoft Game Studios» и «Take-Two Interactive» после 2010 года', fontsize=14, pad=18)
ax.set_xlabel('Объем продаж, млн.', fontsize=12)
ax.set_ylabel('Величина', fontsize=12)
ax.grid(True)
plt.tight_layout()  # Улучшение визуального оформления графиков

plt.show()
```

### Задача 6

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зелёного полупрозрачного прямоугольника.

```python
df_nintendo = df[df['Publisher'] == 'Nintendo']
df_nintendo_year = df_nintendo.groupby('Year')['Name'].count().reset_index()
df_nintendo_year_35 = df_nintendo_year[df_nintendo_year['Name'] > 35]
data = []

fig, ax = plt.subplots(figsize=(12,6))
sns.lineplot(x='Year', y='Name', data=df_nintendo_year, linewidth=2, ax=ax)

ax.grid(True)
# Отрисовка зеленого прямоугольника на нужном промежутке времени
for x1, x2 in zip(df_nintendo_year_35['Year'], df_nintendo_year_35['Year'].shift(-1)):
    if x2 - x1 == 1:
        ax.axvspan(x1, x2, color='green', alpha=0.2)
        data.append(x1)

ax.annotate('больше 35 игр',
            xy=(data[-1], 30),
            xytext=(data[-1]+3, 50),
            arrowprops=dict(
                facecolor='black', # не могу понять, почему стрелка белая остается
                arrowstyle='->',
                connectionstyle="angle3,angleA=-30,angleB=30"
            ),
            fontsize=14,
            color='green')


ax.set_title('Динамика выпуска игр Nintendo по годам', fontsize=14, pad=20)
ax.set_xlabel('Год', fontsize=12)
ax.set_ylabel('Количество игр, шт.', fontsize=12)
plt.tight_layout()  # Улучшение визуального оформления графиков

plt.show()
```

### Задача 7

Определите 3 жанра и 4 платформы с самыми большими продажами игр в мире за всё время. Постройте сетку графиков 4 на 3, в каждой ячейке постройте точечный график, проверяющий зависимость общемировых продаж от продаж в Северной Америке для каждой пары (жанра, платформы). Разными цветами укажите игры разных годов. Размер итогового полотна определите самостоятельно таким образом, чтобы все графики были видны.

Делайте графики достаточно читаемыми, чтобы из них можно было сделать вывод о наличии / отсутствии зависимостей в данных.

```python
top_genres = df.groupby('Genre')['Global_Sales'].sum().nlargest(3).index.tolist()
top_platforms = df.groupby('Platform')['Global_Sales'].sum().nlargest(4).index.tolist()

data = df[(df['Genre'].isin(top_genres)) & (df['Platform'].isin(top_platforms))]

# Создаем сетку графиков
g = sns.relplot(x='NA_Sales', y='Global_Sales', hue='Year', col='Platform', row='Genre', data=data, height=3, aspect=1.5)

# Настраиваем отображение графиков
g.set_axis_labels('Продажи в Северной Америке, млн копий', 'Общие продажи по миру\n, млн копий')
g.fig.suptitle('Зависимость общих продаж от продаж в Северной Америке для топ-жанров и топ-платформ', size=20)

# Отображаем графики
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.3, top=0.9, bottom=0.1, left=0.1, right=0.9)
plt.show()
```
