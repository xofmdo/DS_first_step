### Задача 1

Постройте гистограмму распределения объема продаж всех игр в Японии.

```python
df1 = df[df['JP_Sales'] > 0]['JP_Sales']

fig, ax = plt.subplots(figsize=(12, 7)) # создается объяект Figure и Axes

# тут подробнее, что это такое 
# https://pyprog.pro/mpl/mpl_main_components.html?ysclid=lefw28epht793271838

col_map = plt.get_cmap('Paired') # получаю какой-то список с цветами

_, bins, _  = plt.hist(df1,
         bins = 10, # задается кол-во баров
         edgecolor='black', # задается цвет контуров баров
         color='#FFA500'  # задается цвет баров
) # строю гистограмму

plt.grid(True) # задаю разметку на графике

# Настройка осей
plt.xlabel('Год', fontsize=18, ) # задается подпись и ее размер на оси Х
plt.ylabel('Объем продаж игр в Японии, млн. копий',  fontsize=18)  # задается подпись и ее размер на оси Y
plt.title('Распределения объема продаж всех игр в Японии', fontsize=20, pad=20) # задается подпись и ее размер названия
plt.margins(0.01)

plt.xticks(rotation=0) # так можно задать поворот подписей на оси Х 

plt.show() # так отображаем график
```

```python
import numpy as np

df1 = df[df['JP_Sales'] > 0]['JP_Sales'].apply(np.log)
df2 = df[df['EU_Sales'] > 0]['EU_Sales'].apply(np.log)

fig, ax = plt.subplots(figsize=(12, 7))

col_map = plt.get_cmap('Paired') # получаю какой-то список с цветами

_, bins, _  = plt.hist(df1,
         bins = 20,
         edgecolor='black', 
         color='red', alpha= 0.5) # alpha задает прозрачность

# получаем bins и передаем его в нижнюю гистограмму

plt.hist(df2,
         bins = bins, 
         edgecolor='black', 
         color='green', alpha= 0.5) # передаем bins, чтобы графики были одной ширины


plt.grid(True) # задаю разметку

# Настройка осей
plt.xlabel('Год', fontsize=18, )
plt.ylabel('Объем продаж игр в Японии, млн. копий',  fontsize=18) # fontsize отвечает за размер
plt.title('Распределения объема продаж всех игр в Японии', fontsize=20, pad=20) # pad отвечает за отступ

plt.margins(0.01)  # margins отвечает за отступы от оси Х и Y

ax.tick_params(axis='x', rotation=90) # правильно задаем поворот текста на оси X
ax.set_xticks(list(bins)) # устанавливаем маркеры на оси X

plt.show()
```

### Задача 2

Постройте столбчатую диаграмму (barchart), показывающую сравнение общемирового объема продаж игры "Grand Theft Auto V" на разных платформах.

```python
# готовлю данные
df1 = df[df['Name'] == 'Grand Theft Auto: San Andreas'].groupby('Platform').sum()['Global_Sales'].sort_values()
xlabel = df1.index
value = df1
col_map = plt.get_cmap('Paired') # получаю какой-то список с цветами

fig, ax = plt.subplots(figsize=(12, 7))

plt.grid(True) # задаю разметку

# Настройка осей
ax.bar(xlabel, value, color=col_map.colors, edgecolor='olive', linewidth=1) # устанавливаю подпись, цвет, огранку, ширину бара

# Настройка осей
plt.xlabel('Платформа', fontsize=18, )
plt.ylabel('Объем продаж игр, млн. копий',  fontsize=18)
plt.title('Распределения объема продаж по миру игры\n Grand Theft Auto: San Andreas', fontsize=20, pad=20)
plt.margins(0.01)

plt.xticks(size=16) # задаю размер подписей по оси Х
plt.yticks(size=14)
ax.bar_label(ax.containers[0], padding=3, size=14, rotation=0) # ax.containers[0] достает значение над баром 
# и подписываю значения, размер и поворот баров

ax.set_ylim(-0.5, 22) # задаю границы значений по оси Y 

plt.show()
```

### Задача 4

Постройте точечную диаграмму, показывающую существование зависимости продаж во всем мире от продаж в Северной Америке.

```python
corr = df['Global_Sales'].corr(df['NA_Sales']) # считаю корреляцию
x = df['NA_Sales']
y =  df['Global_Sales']

fig, ax = plt.subplots(figsize=(12, 7))

plt.scatter(x,y, color='black', s=150) # строю точечный график
plt.grid(True) # задаю разметку

text = f"Коэффициент корреляции: {corr:.2f}" # готовлю текст через f-строку

sns.regplot(ax=ax,  # выбираю для какого графика строю (их может быть несколько, но к этому потом)
            data=df, # передаю наш dataframe
            x='NA_Sales', 
            y='Global_Sales', 
            logx=False, #logx - отвечает за то, будет ли линия логарифмированная
            line_kws={'linestyle': '--', 'color': 'black'}) # строю эту линию, задаю данные по x и y, 
    


ax.text(0.75, 
        0.1, 
        text, 
        transform=ax.transAxes,
        ha='center', 
        size=13, 
        color='black', 
        style='italic') # задаю параметры подписи, сначала ее расположение на графике

# Настройка осей
plt.xlabel('Объем продаж игр в Северной Америке,\n млн. копий', fontsize=18, )
plt.ylabel('Объем продаж игр во всем мире, млн. копий',  fontsize=18)
plt.title('Динамики зависимости продаж во всем мире\n от продаж в Северной Америке', fontsize=20, pad=20)
plt.margins(0.01)
plt.xticks(size=16)
plt.yticks(size=14)


plt.show()
```

### Задача 5

Постройте круговую диаграмму процентного соотношения продаж игры "Super Mario Bros." на разных платформах во всем мире. Подпишите доли графика с точностью 1 знак после запятой.

```python
# готовлю данные
sizes = df[df['Name'] == 'Super Mario Bros.'].groupby('Platform').sum()['Global_Sales'].tolist()
labels = df[df['Name'] == 'Super Mario Bros.'].groupby('Platform').sum()['Global_Sales'].index.tolist()

fig, ax = plt.subplots(figsize=(8,8))
explode = [0.07] * len(labels) # задаю параметры отступов внутри круговой диаграммы

ax.pie(sizes, 
       explode=explode, # передаю данные
       #labels=labels, 
       autopct='%1.1f%%', # тут задается, как выводятся подписи
       shadow=True, # будет ли тень под графиком
       startangle=45, # угол поворота диаграммы
       pctdistance = 1.1,
       textprops={'fontsize': 12, 
                  'color': 'red', 
                  'style':'italic', 
                  'fontname':"Times New Roman",
                  'fontweight':"bold"
                  }, # задает параметры для подписи значений

       wedgeprops={'linewidth': 2, 
                   'edgecolor': "olive"}   # задает окантовку
      )

plt.grid(True) # задаю разметку

# Настройка осей
ax.set_title('Распределения объема продаж по миру игры\n Super Mario Bros.', 
             color='black', 
             style='italic',
             fontname="Times New Roman", 
             size=16,
             fontweight="bold") # подписываю название
plt.margins(0.01)
plt.xticks(size=16)
plt.yticks(size=14)

ax.legend(labels, loc='upper right', bbox_to_anchor=(1.2, 1)) # наносим легенду на график 


plt.show()
```

Одной из основных задач при анализе данных является проверка данных на корректность и отсутствие ошибок. В датасете "vgsales" одним из потенциальных источников ошибок является столбец с общемировыми продажами. По идее, значение общемировых продаж какой-либо игры должно равняться сумме продаж в Северной Америке, Европе, Японии и остальных странах, в противном случае мы получим ошибку в данных, которая может негативно повлиять на весь дальнейший процесс анализа.

Причин возникновения такой ошибки может быть несколько. Давайте рассмотрим следующие причины:
1. Ошибка округления. В целом, такая ошибка достаточно часто встречается при переводе величин из одних единиц в другие (например, из тысяч копий в миллионы копий).
1. Ошибка вычислений. Такая ситуация требует более серьезного исследования, поскольку она может привести к недостоверным результатам анализа данных.

Будем считать, что в записи имеется ошибка суммирования, если сумма продаж какой-либо игры в Северной Америке, Европе, Японии и остальных странах отличается от общемировых продаж более, чем на 0.01 млн копий.

С помощью круговой диаграммы проиллюстрируйте процент записей в датасете, имеющих такую ошибку суммирования продаж.

```python
df.head()
```

```python
df['my_summary'] = (df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales']+ df['Other_Sales'] - df['Global_Sales']).abs() > 0.01

data = df['my_summary'].value_counts()
data
```

```python
sizes = data.tolist()
labels = ['Не имеют ошибок', 'Имеют ошибки'] # задаю лейблы

fig, ax = plt.subplots(figsize=(8,8))
explode = [0.07] * len(labels)

ax.pie(sizes, 
       explode=explode,
       #labels=labels, 
       autopct='%1.2f%%',
       shadow=True, 
       startangle=45,
       pctdistance = 1.1,
       textprops={'fontsize': 12, 
                  'color': 'black', 
                  'style':'italic', 
                  'fontname':"Times New Roman",
                  'fontweight':"bold"
                  }, 

       wedgeprops={'linewidth': 1, 
                   'edgecolor': "black"}
      )

plt.grid(True) # задаю разметку

# Настройка осей
ax.set_title('Процент записей в датасете, имеющих\n ошибку суммирования продаж.', 
             color='black', 
             style='italic',
             fontname="Times New Roman", 
             size=16,
             fontweight="bold")
plt.margins(0.01)
plt.xticks(size=16)
plt.yticks(size=14)

ax.legend(labels, loc='upper right', bbox_to_anchor=(1.2, 1))


plt.show()
```
